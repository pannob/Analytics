```python
# I have created this function to (first line) keep best grades and (second line) to replace missing grades with 0

def tidy_grades(x):
    x = x.drop_duplicates(subset=('GRADE','RESEARCH_ID')).groupby(['RESEARCH_ID'], as_index=False).max()
    x.replace('-', 0.00, inplace = True)
    return x
```

```python
# creating a generator that makes random numbers

rng = random.default_rng()
```

```python
# creating a variable list of 20 random numbers from 0 to 150

randomnumbers = rng.integers(0, 150, size = 20)
```

```python
randomnumbers
```

```python
# using my above variable as indices to slice the dataframe

# WARNING - there is a chance that one of the random numbers selected is not a RESEARCH_ID, re-run code to select new #s

newset = dfsumtestFormattedRNG.loc[randomnumbers]
```

```python
newset
```

```python
# replacing the grades in q1,2,3 with random number across the sliced dataset

newset.loc[:,'Q1'] = np.random.random(size=len(newset))
newset.loc[:,'Q2'] = np.random.random(size=len(newset))
newset.loc[:,'Q3'] = np.random.random(size=len(newset))
```

```python
newsettidy = round_decimals(newset)
```

```python
newsettidy
```

```python
# recalculating the grade column by a cumulative mean including updated q1,2,3 

newsettidy['GRADE'] =  newsettidy[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 
                                   'Q7', 'Q8', 'Q9', 'Q10', 'Q11', 'Q12', 'Q13']].mean(axis=1)
```

```python
dfsumtestsubset = round_decimals(newsettidy)
```

```python
dfsumtestsubset
```

```python
dfsumtestFormattedRNG.loc[randomnumbers] = dfsumtestsubset
```
